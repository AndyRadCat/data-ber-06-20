```python
data_movies['reviews_from_users'] = data_movies['reviews_from_users'].fillna(0)
data_movies['reviews_from_users'] = data_movies['reviews_from_users'].astype(int) 
data_movies.dtypes
```

```python
data_movies['reviews_from_critics'] = data_movies['reviews_from_critics'].fillna(0)
data_movies['reviews_from_critics'] = data_movies['reviews_from_critics'].astype(int) 
```

```python
data_movies['usa_gross_income'] = data_movies['usa_gross_income'].str.replace('$','')
data_movies['usa_gross_income'] = data_movies['usa_gross_income'].str.replace('EUR','')
data_movies['usa_gross_income'] = data_movies['usa_gross_income'].str.replace('GBP','')
data_movies['usa_gross_income'] = data_movies['usa_gross_income'].str.replace('HKD','')
```

```python
data_movies['usa_gross_income'] = data_movies['usa_gross_income'].fillna(0)
data_movies['usa_gross_income'] = data_movies['usa_gross_income'].astype(int) 
```
